```python
import pandas as pd
import numpy as np

customers = pd.DataFrame({
    "Customer ID": df["Customer ID"].unique(),
    "Customer Name": np.random.choice(
        ["Amit", "Neha", "Raj", "Simran", None],
        size=len(df["Customer ID"].unique())
    ),
    "Segment": np.random.choice(
        ["Consumer", "Corporate", "Home Office", None],
        size=len(df["Customer ID"].unique())
    )
})

customers.to_csv("../data/raw/customers_raw.csv", index=False)
```

```python
products = pd.DataFrame({
    "Product ID": df["Product ID"].unique(),
    "Category": np.random.choice(
        ["Furniture", "Office Supplies", "Technology"],
        size=len(df["Product ID"].unique())
    ),
    "Sub-Category": np.random.choice(
        ["Chairs", "Tables", "Phones", "Binders", None],
        size=len(df["Product ID"].unique())
    )
})

products.to_csv("../data/raw/products_raw.csv", index=False)
```

```python
regions = pd.DataFrame({
    "region_id": ["R1", "R2", "R3", "R4"],
    "region_name": ["Central", "East", "South", "West"],
    "manager": ["Anil", None, "Pooja", "Rahul"]
})

regions.to_csv("../data/raw/regions_raw.csv", index=False)
```

```python
inventory = pd.DataFrame({
    "Product ID": df["Product ID"].unique(),
    "stock_level": np.random.randint(-10, 500, len(df["Product ID"].unique())),
    "last_restock_date": pd.to_datetime("2023-01-01") +
        pd.to_timedelta(
            np.random.randint(0, 180, len(df["Product ID"].unique())),
            unit="D"
        )
})

inventory.to_csv("../data/raw/inventory_raw.csv", index=False)
```
